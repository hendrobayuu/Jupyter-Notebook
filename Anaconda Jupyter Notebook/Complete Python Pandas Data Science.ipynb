# Regex Filtering (filter based on textual patterns)

```python
data_frame.loc[data_frame['Name'].str.contains('Mega')] #filtering berdasarkan nama dengan kandungan kata Mega
```

```python
data_frame.loc[~data_frame['Name'].str.contains('Mega')] #filtering berdasarkan nama dengan tidak terkandungan kata Mega
```

```python
import re #library regular expression (regex) deretan karakter yang digunakan untuk pencarian string atau teks dengan menggunakan pola (pattern)
data_frame.loc[data_frame['Type 1'].str.contains('Fire|Grass', regex=True)] #filtering berdasarkan regular expression nama Fire dan Grass pada Type 1
```

```python
data_frame.loc[data_frame['Type 1'].str.contains('fire|grass',flags=re.I, regex=True)] #Sama aja fungsinya dengan di atas, tapi terlihat berbeda bukan
```

```python
data_frame.loc[data_frame['Name'].str.contains('^pi[a-z]*',flags=re.I, regex=True)] #filtering berdasarkan kondisi Name dengan huruf depan pi
```

# Conditional Changes

```python
#data_frame.loc[data_frame['Type 1'] == 'Fire', 'Type 1'] = 'Flammer' #mengubah kata fire menjadi flammer pada Type 1
data_frame.loc[data_frame['Type 1'] == 'Flammer', 'Type 1'] = 'Fire' #mengubah kata flammer menjadi fire pada Type 1

data_frame
```

```python
data_frame.loc[data_frame['Type 1'] == 'Fire', 'Legendary'] = True #mengubah kata fire menjadi flammer pada Type 1
data_frame
```

```python
data_frame.loc[data_frame['Total']>500, ['Generation', 'Legendary']] = 'CHANGE' #mengubah setiap total>500 pada kolom generation dan kolom legendary menjadi terisi kata CHANGE
data_frame
```

```python
data_frame.loc[data_frame['Total']>500, ['Generation', 'Legendary']] =['CHANGE-1', 'CHANGE 2'] #mengubah setiap total>500 pada kolom generation dan kolom legendary menjadi terisi kata CHANGE-1 LALU CHANGE-2
data_frame
```

# Aggregate statistic using groupby (sum, mean, counting)

```python
data_frame.groupby(['Type 1']).mean() #menampilkan rata-rata denan berdasarkan tipe di Type 1
```

```python
data_frame.groupby(['Type 1']).mean().sort_values('Defense', ascending = True) #menampilkan rata-rata denan berdasarkan tipe di Type 1 dan mengurutkan berdasrkan defense dari rendah ke tinggi
#.sum() .count()
```

```python
data_frame['count']=1 #terlihat bedanya bukan dengan yang di atas?
data_frame.groupby(['Type 1']).count()['count']
```
